```
import pandas as pd
import numpy as np
```

```
df_sumarized = pd.read_excel('nasa_1.xlsx')
```

```
df_index = pd.read_csv("gender-development-index.csv")
```

# DATA PREPARATION

```
import seaborn as sns
import matplotlib.pyplot as plt

df_numeric = df_sumarized.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
```

```
df_sumarized = df_sumarized.drop(columns=["Buffalo Stocks (1000s) 1994", "Camel Stocks (1000s) 1994"])
```

```
latin_america_countries = [
    "Argentina", "Belize", "Bolivia", "Brazil", "Chile", "Colombia",
    "Costa Rica", "Ecuador", "El Salvador", "French Guiana", "Guatemala",
    "Guyana", "Honduras", "Mexico", "Nicaragua", "Panama",
    "Paraguay", "Peru", "Suriname", "Uruguay", "Venezuela"
]
```

```
df_gender = df_index[df_index['Entity'].isin(latin_america_countries)]
```

```
df_gender_by_country = df_gender.groupby('Entity')['Gender Development Index'].mean().reset_index()
```

```
df_summerized_gender = pd.merge(df_sumarized, df_gender_by_country, left_on='Country', right_on='Entity', how='left')
```

#  M0DELING - Gender equality_Socioeconomics

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df_numeric = df_summerized_gender.drop(columns=['Entity', 'Country']).apply(pd.to_numeric, errors='coerce')

correlation_matrix = df_numeric.corr()

correlaciones_gdi = correlation_matrix['Gender Development Index'].sort_values()

mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(24, 16))

sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', linewidths=0.5)

# Mostrar el gr√°fico
plt.title("Correlation Matrix - Gender Development Index")
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_numeric = df_summerized_gender.drop(columns=['Entity', 'Country']).apply(pd.to_numeric, errors='coerce')

correlation_matrix = df_numeric.corr()

correlaciones_gdi = correlation_matrix['Gender Development Index'].sort_values()

top_positive = correlaciones_gdi.tail(11)
top_negative = correlaciones_gdi.head(10)

top_correlations = pd.concat([top_negative, top_positive])

plt.figure(figsize=(10, 6))
sns.barplot(x=top_correlations.values, y=top_correlations.index, palette='coolwarm')
plt.title('Top 10 most positive and negative variables regarding Gender Development Index')
plt.xlabel('Correlation')
plt.ylabel('Variables')
plt.axvline(0, color='black', linestyle='--')
plt.show()
```
