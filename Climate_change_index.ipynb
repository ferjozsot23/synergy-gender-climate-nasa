# CLIMATE CHANGE INDEX



# CO2

```python
import pandas as pd
df_co2 = pd.read_csv('co-emissions-per-capita.csv')
df_co2.head()
```

```python
df_co2_by_country = df_co2.groupby('Entity')['Annual CO₂ emissions (per capita)'].mean().reset_index()
```

```python
df_co2_by_country[df_co2_by_country['Entity'] == 'South America']
```

```python
latin_america_countries = [
    "Argentina", "Belize", "Bolivia", "Brazil", "Chile", "Colombia",
    "Costa Rica", "Ecuador", "El Salvador", "French Guiana", "Guatemala",
    "Guyana", "Honduras", "Mexico", "Nicaragua", "Panama",
    "Paraguay", "Peru", "Suriname", "Uruguay", "Venezuela"
]
```

```python
len(latin_america_countries)
```

```python
df_co2_by_country_latin_america = df_co2_by_country[df_co2_by_country['Entity'].isin(latin_america_countries)]
```

```python
df_co2_by_country_latin_america.Entity.unique()
```

```python
df_co2_by_country_latin_america.rename(columns={'Annual CO₂ emissions (per capita)': 'CO2'}, inplace=True)
```

# FOSSIL

```python
df_fossil = pd.read_csv('fossil-fuels-per-capita.csv')
```

```python
df_fossil_by_country = df_fossil.groupby('Entity')['Fossil fuels per capita (kWh)'].mean().reset_index()
```

```python
df_fossil_by_country[df_fossil_by_country['Entity'] == 'South America']
```

```python
df_fossil_by_country_latin_america = df_fossil_by_country[df_fossil_by_country['Entity'].isin(latin_america_countries)]
```

```python
df_fossil_by_country_latin_america.rename(columns={'Fossil fuels per capita (kWh)': 'Fossil'}, inplace=True)
```

```python
df_fossil_by_country_latin_america
```

# Renewable energy

```python
df_renewable = pd.read_csv('modern-renewable-energy-consumption.csv')
```

```python
import pandas as pd
df_renewable_by_country = df_renewable.groupby('Entity')[[
    'Other renewables (including geothermal and biomass) electricity generation - TWh',
    'Solar generation - TWh',
    'Wind generation - TWh',
    'Hydro generation - TWh'
]].mean().reset_index()

df_renewable_by_country.columns = [
    'Entity',
    'Mean Other Renewables - TWh',
    'Mean Solar Generation - TWh',
    'Mean Wind Generation - TWh',
    'Mean Hydro Generation - TWh'
]
df_renewable_by_country['Mean Total Renewable Generation - TWh'] = df_renewable_by_country[['Mean Other Renewables - TWh',
                                                                                               'Mean Solar Generation - TWh',
                                                                                               'Mean Wind Generation - TWh',
                                                                                               'Mean Hydro Generation - TWh']].mean(axis=1)
df_renewable_by_country
```

```python
df_renewable_by_country_latin_america = df_renewable_by_country[df_renewable_by_country['Entity'].isin(latin_america_countries)]
```

```python
df_renewable_by_country_latin_america.rename(columns={'Mean Total Renewable Generation - TWh': 'Renewable'}, inplace=True)
```

```python
df_renewable_by_country_latin_america = df_renewable_by_country_latin_america[['Entity', 'Renewable']]
```

# ENERGY

```python
df_energy = pd.read_csv('per-capita-energy-use.csv')
```

```python
df_energy_by_country = df_energy.groupby('Entity')['Primary energy consumption per capita (kWh/person)'].mean().reset_index()
```

```python
df_energy_by_country[df_energy_by_country['Entity'] == 'South America']
```

```python
df_energy_by_country_latin_america = df_energy_by_country[df_energy_by_country['Entity'].isin(latin_america_countries)]
```

```python
df_energy_by_country_latin_america.rename(columns={'Primary energy consumption per capita (kWh/person)': 'Energy'}, inplace=True)
```

```python
df_combined = df_co2_by_country_latin_america.merge(df_fossil_by_country_latin_america, on='Entity', how='outer', suffixes=('', '_fossil'))
df_combined = df_combined.merge(df_renewable_by_country_latin_america, on='Entity', how='outer', suffixes=('', '_renewable'))
df_combined = df_combined.merge(df_energy_by_country_latin_america, on='Entity', how='outer', suffixes=('', '_energy'))
```

```python
df_cc_index = df_combined.copy()
```

```python
columns = {
    'CO2': 'CO2',
    'Fossil': 'Fossil',
    'Renewable': 'Renewable',
    'Energy': 'Energy'
}

df_cc_index_normalized = df_cc_index.copy()

for col in columns.values():
    df_cc_index_normalized[col] = (df_cc_index[col] - df_cc_index[col].min()) / (df_cc_index[col].max() - df_cc_index[col].min())

df_cc_index_normalized
```

```python
df_cc_index_normalized.drop(columns=['Fossil', 'Renewable'], inplace=True)
```

```python
df_cc_index_normalized.head()
```

```python
df_cc_index_normalized['ccindex'] = (df_cc_index_normalized['CO2'] * 0.6) + (df_cc_index_normalized['Energy'] * 0.4)
```

```python
df_cc_index_normalized
```

```python
df_cci = df_cc_index_normalized[['Entity', 'ccindex']]
```
